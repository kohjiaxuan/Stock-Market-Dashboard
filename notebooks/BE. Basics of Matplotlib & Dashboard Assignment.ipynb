```python
import matplotlib.pyplot as plt
#import submodule pyplot from matplotlib

#For Python free version, want to display chart in the same diagram
#Create chart 'in line' and not a new window
%matplotlib inline
```

```python
#prices of apple and microsoft over mid Nov of each year

apl_price = [93.95,112.15,104.05,144.85,169.49]
ms_price = [39.01,50.29,57.05,69.98,94.39]
year = [2014, 2015, 2016, 2017, 2018]
```

```python
#Easiest way to plot quick charts
plt.plot(year,apl_price)
plt.plot(year, ms_price)
plt.xlabel('Year')
plt.ylabel('Stock Price')
#Show graph in first run, not second
plt.show()
```

```python
#Make scatter plot, use plt.scatter instead of plt.plot
plt.scatter(year,apl_price)
plt.scatter(year, ms_price)
plt.xlabel('Year')
plt.ylabel('Stock Price')
plt.show()
```

```python
#Plotting multiple plots in single line of plt.plot
#plt.plot(x1,y1,x2,y2,...)
plt.plot(year,apl_price,
         year, ms_price)

plt.xlabel('Year')
plt.ylabel('Stock Price')
plt.show()
```

```python
#Changing line colour
#k for black, g for green
plt.plot(year,apl_price, 'k',
         year, ms_price, 'g')

plt.xlabel('Year')
plt.ylabel('Stock Price')
plt.show()
```

```python
#Changing line style
#'-', '--', '-.', ':', 'steps' etc
#Default is '-' for straight line
plt.plot(year, apl_price, '-',
         year, ms_price, '-.')

plt.xlabel('Year')
plt.ylabel('Stock Price')
plt.show()
```

```python
#Changing line style + line colour + dot style
#In 3rd and 6th argument corresponding to 1st and 2nd graph
#1 Comment argument with ''
#2 Graph type e.g. -- or :,
#3 Colour of line b-blue, c-cyan etc.
#4 Type of dot, e.g. x-cross, o-dot
plt.plot(year,apl_price, '--bx',
         year, ms_price, ':co')

plt.xlabel('Year')
plt.ylabel('Stock Price')
plt.show()
```

```python
plt.plot(year,apl_price, '--ok', year, ms_price, ':r.')
# '--ok' means dotted graph --, solid dot o for point and k is black
# ':r.' means : for fine dotted graph, r for red and . for small dot
#Note how order of line style, dot style and line colour dont matter
plt.xlabel('Year')
plt.ylabel('Stock Price')

#Change axis
#plt.axis([xmin, xmax, ymin, ymax])
plt.axis([2013, 2019, 20, 180])

plt.show()
```

```python
#Manual way of creating plots, has more control

#matplotlib is 3 layered, figure, subplot/chart in subplot and axis
#figure is space where all charts are shown (often in a grid/matrix)
#figure can be divided to mxn matrix space
#chart is put inside a cell of the matrix space of the figure
#axis is the axis?


#Create a new figure in matplotlib using plt.figure()
#Save into variable fig_1 for future use

fig_1 = plt.figure(1,figsize=(12,4.8))
#Define space given for figure
#plt.figure(fig attribute int,figsize=(width-inches,height-inches))

#Add charts (create chart1, chart2 var) to figure (fig_1 var)
#Charts are put into spaces in a figure called subplot
#.add_subplot(totalrowno.totalcolno.positionnumberofchart)
#.add_subplot(121) means figure divided into 1x2 matrix, graph at position 1
#.add_subplot(122) means figure divided into 1x2 matrix, graph at position 2
#1x2 matrix can have 2 graphs, going from top left, left to right
#Figure matrix/grid layout for graph
#1 2
chart_1 = fig_1.add_subplot(121)
chart_2 = fig_1.add_subplot(122)

chart_1.plot(year,apl_price, '-rx')
chart_2.scatter(year,ms_price)

plt.show()
```

```python
from matplotlib.ticker import MaxNLocator
#Tick locating - locators handle autoscaling of view limits based on data limits
#and choosing of tick locations

#Useful semi-automatic tick locator is MultipleLocator
#initialized with a base e.g. 10 and picks axis limits/ticks that are multiples of that base
#Or you can ue MaxNLocator, finds up to max number of intervals with ticks at nice location
#AutoLocator is MaxNLocator with simple defaults
#Others: LinearLocator (evenly spaced ticks), LogLocator etc.
#You can import any of these for use from matplotlib

fig_1 = plt.figure(1,figsize=(12,4.8))

apple_price = fig_1.add_subplot(121)
microsoft_price = fig_1.add_subplot(122)

#Use tick locator
#Besides finding right number/places for axis ticks, also can change formatting to integer
apple_price.xaxis.set_major_locator(MaxNLocator(integer=True))
microsoft_price.xaxis.set_major_locator(MaxNLocator(integer=True))

apple_price.plot(year,apl_price, '-rx')
microsoft_price.scatter(year,ms_price)

plt.show()
```

```python
#If we have 3 or 4 graph, we need a 2x2 subplot style
fig_2 = plt.figure(1,figsize=(12,10))

#Since subplots can overlay each other
#Create a big subplot of 1x1 matrix taking up entire figure
#Set title using this subplot and words will appear on top

#frameon=False hides 4 sides of the frame of title
titlebigspace = fig_2.add_subplot(111, frameon=False)

#Chart subplot must be defined later than titlebigspace
#Because the later subplots cover the earlier ones
#Be careful to reference right figure when adding subplot
chart2_1 = fig_2.add_subplot(221)
chart2_2 = fig_2.add_subplot(222)

#Add plots to subplot
chart2_1.plot(year,apl_price, '-rx')
chart2_2.scatter(year,ms_price)

#Add Amazon plot to 3rd subplot
amazon_price = [332.63,668.45,760.16,1186,1502.06]
chart2_3 = fig_2.add_subplot(223)
chart2_3.plot(year,amazon_price,'g--')

#Set labels of each variable representing a subplot with graph
#instead of plt.xlabel or plt.title for current graph, use var.set_xlabel, var.set_ylabel, var.set_title
chart2_1.set_xlabel('Year')
chart2_1.set_ylabel('Apple Price')
#Commented title of Apple/Microsoft as it blocks main title
#chart2_1.set_title('Stock Price of Apple')
chart2_2.set_xlabel('Year')
chart2_2.set_ylabel('Microsoft Price')
#chart2_2.set_title('Stock Price of Microsoft')
chart2_3.set_xlabel('Year')
chart2_3.set_ylabel('Amazon Price')
chart2_3.set_title('Stock Price of Amazon')


# Turn off axis lines and ticks of the big subplot to not block other graphs
#Can use frameon=False instead of setting colour of spines to 'none'
#titlebigspace.spines['top'].set_color('none')
#titlebigspace.spines['bottom'].set_color('none')
#titlebigspace.spines['left'].set_color('none')
#titlebigspace.spines['right'].set_color('none')

#Remove tick parameters that block the graph as titlebigspace is just used for title
titlebigspace.tick_params(labelcolor='none', top='off', bottom='off', left='off', right='off')

#Set title/xlabel of big space that represents all smaller charts
titlebigspace.set_title('Important stock prices of IT companies')
titlebigspace.set_xlabel('From 2013-2018')

#Convert axis ticks formatting to integer and smartly determine ticks to show/tick interval
chart2_1.xaxis.set_major_locator(MaxNLocator(integer=True))
chart2_2.xaxis.set_major_locator(MaxNLocator(integer=True))
chart2_3.xaxis.set_major_locator(MaxNLocator(integer=True))

plt.show()

#Important to put all graphs inside before showing as there are some issues
#with showing graphs after plt.show() and adding new subplots
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator

#Assignment: Plot a graph of 9 companies in a 6x3 plot (using top 3x3)
#And for bottom 3x3, plot all 9 graphs together with normalized value from initial year 2014
#Range from 2014 to 2018
year = [2014, 2015, 2016, 2017, 2018]

#Create list of all 9 companies and the years from end Nov each year 2014-2018
apple_price = [93.95,112.15,104.05,144.85,169.49]
microsoft_price = [39.01,50.29,57.05,69.98,94.39]
amazon_price = [332.63,668.45,760.16,1186,1502.06]
google_price = [523.52,758.88,771.82,1046.40,1035.61]
samsung_price = [24460/1150,25700/1150,31720/1150,55460/1150,42400/1150]
facebook_price = [73.75,107.32,117.02,182.78,131.73]
netflix_price = [51.47,123.84,115.21,195.75,258.82]
huawei_price = [6.95/6.5,13.2/6.5,13.80/6.5,8.99/6.5,4.67/6.5]
alibaba_price = [110.73,79.95,93.39,191.19,150.33]


#Create normalized list of 9 companies stock prices
apple_price_n = np.array(apple_price)/apple_price[0]
microsoft_price_n = np.array(microsoft_price)/microsoft_price[0]
amazon_price_n = np.array(amazon_price)/amazon_price[0]
google_price_n = np.array(google_price)/google_price[0]
samsung_price_n = np.array(samsung_price)/samsung_price[0]
facebook_price_n = np.array(facebook_price)/facebook_price[0]
netflix_price_n = np.array(netflix_price)/netflix_price[0]
huawei_price_n = np.array(huawei_price)/huawei_price[0]
alibaba_price_n = np.array(alibaba_price)/alibaba_price[0]

#Plan of subplot layering (1x1 one for title, 2x1 for overlaid graphs, 6x3 one for small graphs but only occupy top 9 spaces)

#Create big figure
stocks = plt.figure(1,figsize=(24,30))

#Create big subplot for title and remove frame (frameon=False), remove tick parameters
stocks_title = stocks.add_subplot(111, frameon=False) #remove frame but need remove ticks/axes
stocks_title.tick_params(labelcolor='none', top='off', bottom='off', left='off', right='off')
stocks_title.set_title('Graph of stock prices (USD) from 2014-2018 of 9 tech companies', fontsize=32)

#Create big subplot for mega chart using 2x1 plot using normal .add_subplot
#Change x-axis to be integer (year) and y-axis to be integer (price_index)
stocks_big = stocks.add_subplot(212)
apple_n = stocks_big.plot(year, apple_price_n, '-o', label='Apple', linewidth = 4, markersize=8)
microsoft_n = stocks_big.plot(year, microsoft_price_n, '-o', label='Microsoft', linewidth = 4, markersize=8)
amazon_n = stocks_big.plot(year, amazon_price_n, '-o', label='Amazon', linewidth = 4, markersize=8)
google_n = stocks_big.plot(year, google_price_n, '-o', label='Google', linewidth = 4, markersize=8)
samsung_n = stocks_big.plot(year, samsung_price_n, '-o', label='Samsung', linewidth = 4, markersize=8)
facebook_n = stocks_big.plot(year, facebook_price_n, '-o', label='Facebook', linewidth = 4, markersize=8)
netflix_n = stocks_big.plot(year, netflix_price_n, '-o', label='Netflix', linewidth = 4, markersize=8)
huawei_n = stocks_big.plot(year, huawei_price_n, '-o', label='Huawei', linewidth = 4, markersize=8)
alibaba_n = stocks_big.plot(year, alibaba_price_n, '-o', label='Alibaba', linewidth = 4, markersize=8)
#legend_n = stocks_big.legend(loc=2, fontsize=24) #see below for loc positions in subplot

#get all handle/variable storing the legend in a tuple from 0-8 and also all label of the legend var in a tuple from 0-8
#each element n of the tuple has a pair of handle/label associated
handlesN, labelsN = plt.gca().get_legend_handles_labels()
order = [6,2,1,3,0,5,4,8,7] #labels checked for order of best performing stock and least performing n index auto obtained
stocks_big.legend([handlesN[idx] for idx in order],[labelsN[idx] for idx in order], loc=2, fontsize=24)
stocks_big.tick_params(axis='both', which='major', labelsize=22)
stocks_big.set_title('Graph of stock price index of all major tech companies', fontsize=26)
stocks_big.set_xlabel('Year',fontsize=24)
stocks_big.set_ylabel('Stock Price Index',fontsize=24)
stocks_big.xaxis.set_major_locator(MaxNLocator(integer=True))
stocks_big.yaxis.set_major_locator(MaxNLocator(integer=True))

#Create figure with subplot size 6x3 plot using fig, axes_small (axes_small is np.array)
#Found out that it is easier to use big figure and overlay many different size of subplot (1x1 title), (2x1 big chart),
#(6x3 small charts) into a single figure
#instead of using fig, axes to define fig and axes (matrix size) tgt since there is varying sizes

#Change x-axis to be integer (year) and y-axis to be float (prices)
netflix_s = stocks.add_subplot(631)
netflix_s.plot(year,netflix_price, '-o',color=(231/255,84/255,128/255), label='Netflix') #define color as rgb tuple
netflix_s.legend(fontsize=16, loc=2)
netflix_s.tick_params(axis='both', which='major', labelsize=18)
netflix_s.xaxis.set_major_locator(MaxNLocator(integer=True))

amazon_s = stocks.add_subplot(632)
amazon_s.plot(year,amazon_price, '-o', color='green', label='Amazon')
amazon_s.legend(fontsize=16, loc=2)
amazon_s.tick_params(axis='both', which='major', labelsize=14)
amazon_s.xaxis.set_major_locator(MaxNLocator(integer=True))

microsoft_s = stocks.add_subplot(633)
microsoft_s.plot(year,microsoft_price, '-o', color='orange', label='Microsoft')
microsoft_s.legend(fontsize=16, loc=2)
microsoft_s.tick_params(axis='both', which='major', labelsize=14)
microsoft_s.xaxis.set_major_locator(MaxNLocator(integer=True))

google_s = stocks.add_subplot(634)
google_s.plot(year,google_price, '-o', color='red', label='Google')
google_s.legend(fontsize=16, loc=2)
google_s.tick_params(axis='both', which='major', labelsize=14)
google_s.xaxis.set_major_locator(MaxNLocator(integer=True))

apple_s = stocks.add_subplot(635)
apple_s.plot(year,apple_price, '-o', color='blue', label='Apple')
apple_s.legend(fontsize=16, loc=2)
apple_s.tick_params(axis='both', which='major', labelsize=14)
apple_s.xaxis.set_major_locator(MaxNLocator(integer=True))

facebook_s = stocks.add_subplot(636)
facebook_s.plot(year,facebook_price, '-o', color='brown', label='Facebook')
facebook_s.legend(fontsize=16, loc=2)
facebook_s.tick_params(axis='both', which='major', labelsize=14)
facebook_s.xaxis.set_major_locator(MaxNLocator(integer=True))

samsung_s = stocks.add_subplot(637)
samsung_s.plot(year,samsung_price, '-o',color='m', label='Samsung') #m is magenta
samsung_s.legend(fontsize=16, loc=2)
samsung_s.tick_params(axis='both', which='major', labelsize=14)
samsung_s.xaxis.set_major_locator(MaxNLocator(integer=True))

alibaba_s = stocks.add_subplot(638)
alibaba_s.plot(year,alibaba_price, '-o', color='#999900', label='Alibaba') #can put html hex code for dark yellow
alibaba_s.legend(fontsize=16, loc=2)
alibaba_s.tick_params(axis='both', which='major', labelsize=14)
alibaba_s.xaxis.set_major_locator(MaxNLocator(integer=True))

huawei_s = stocks.add_subplot(639)
huawei_s.plot(year,huawei_price, '-o', color='0.25', label='Huawei') #value of 0-1 as string means gray
huawei_s.legend(fontsize=16, loc=1)
huawei_s.tick_params(axis='both', which='major', labelsize=14)
huawei_s.xaxis.set_major_locator(MaxNLocator(integer=True))
#Create all 9 subplots in small cells with ORIGINAL prices

plt.show()
```
